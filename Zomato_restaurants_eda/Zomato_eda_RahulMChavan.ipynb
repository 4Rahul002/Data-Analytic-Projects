Find the most number of restaurants per city

```python
restaurant_count_per_city=df.groupby('city')['res_id'].nunique().sort_values(ascending=False)
top_5_cities=restaurant_count_per_city.head(5)

plt.figure(figsize=(10,6))
plt.grid(True)
ax=sns.barplot(x=top_5_cities.values,y=top_5_cities.index,hue=top_5_cities.index,palette='viridis')

for container in ax.containers:
  ax.bar_label(container,fontsize=10)

plt.title('Top 5 Cities with Most Restaurants',fontsize=16)
plt.xlabel('Number of Restaurants',fontsize=14)
plt.ylabel('City',fontsize=14)
plt.show()

```

Conclude that Most restaurants are set up in Bangalore , Chennai and New Delhi

Find the most number of chains in india

```python
chains.head(5)
```

```python
top_5_chains=chains.head(5)

plt.figure(figsize=(10,6))
plt.grid(True)
ax=sns.barplot(x=top_5_chains.values,y=top_5_chains.index,hue=top_5_chains.index,palette='coolwarm')
plt.title("Top 5 Chain Restaurants in India", fontsize=14)
plt.xlabel("Number of Restaurants", fontsize=12)
plt.ylabel("Restaurant Name", fontsize=12)

for container in ax.containers:
  ax.bar_label(container,fontsize=10)

plt.show()
```

Find out which establishment is the most popular

Now Lets find the average cost for two in each city

```python
city_avg_cost=df.groupby("city")['average_cost_for_two'].mean().sort_values(ascending=False)
top_5_cities=city_avg_cost.head(5)
plt.figure(figsize=(10,6))

ax=sns.barplot(y=top_5_cities.values,x=top_5_cities.index,hue=top_5_cities.index,palette='Set1')
plt.title("Top 5 Cities with Highest Average Cost for Two", fontsize=14)
plt.ylabel("Average Cost for Two", fontsize=12)
plt.xlabel("City", fontsize=12)
plt.xticks(rotation=45,ha='right')

for container in ax.containers:
  ax.bar_label(container,fontsize=10)

plt.show()
```

As the above plot shows we see

Rs.1075 is  Mumbai which is the most expensive in terms of average cost for two.

Now lets find the aggregate_rating per City

```python
average_ratings=df.groupby("city")['aggregate_rating'].mean().round(2).sort_values(ascending=False)
top_5_cities=average_ratings.head(5)
plt.figure(figsize=(15,6))
ax=sns.barplot(x=top_5_cities.values,y=top_5_cities.index,hue=top_5_cities.index,palette='Set2')

for container in ax.containers:
  ax.bar_label(container,fontsize=10)

plt.title("Average Aggregate Rating in each city",fontsize=14)
plt.xlabel("Average Rating",fontsize=12)
plt.ylabel("City",fontsize=12)
plt.show()
```

So , we can conclude that

Gurgaon has Highest rated Restaurants in India.

Bangalore , Mumbai which has high number of restaurants accociated with them , perform not so well on the ratings.

So now lets find the relation between aggregate rating ,vote count and photo count


```python
fig,ax=plt.subplots(1,2,figsize=(15,6))

sns.scatterplot(x="aggregate_rating",y="votes",data=df,ax=ax[0],palette='magma')
ax[0].set_title("Votes vs Aggregate Rating",fontsize=14)
ax[0].set_xlabel("Aggregate Rating",fontsize=12)
ax[0].set_ylabel("votes",fontsize=12)

sns.scatterplot(data=df,x="aggregate_rating",y="photo_count",ax=ax[1],palette='magma')
ax[1].set_title("Photocount vs Aggregate Rating", fontsize=14)
ax[1].set_xlabel("Aggregate Rating",fontsize=12)
ax[1].set_ylabel("Photo Count",fontsize=12)

plt.show()
```

We can conclude from above both plots

Aggregate ratings are mainly 4-5 be the case votes or photocount,

Thus concluding that 4 - 5 star restaurants have more popularity and feedback given to them .

Now Lets explore how price range differs based on votes and photocount

using heatmap

```python
correlation_matrix=df[['price_range','votes','photo_count']].corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Price Range Vs votes Vs photo Count")
plt.show()
```

As you can see

Photo_count and votes are more closely related than the price range - 0.75
