## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Selecting numerical columns for outlier analysis
numerical_columns = ["runtime", "imdb_score", "imdb_votes", "tmdb_popularity", "tmdb_score"]

# Plotting boxplots for outlier detection
plt.figure(figsize=(15, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column}")

plt.tight_layout()
plt.show()
````

````python

numerical_columns = ["runtime", "imdb_score", "imdb_votes", "tmdb_popularity", "tmdb_score"]

# Handling Skewness: Apply Log Transformation to Highly Skewed Columns
skewed_columns = ["imdb_votes", "tmdb_popularity"]
for col in skewed_columns:
    df[col] = np.log1p(df[col])  # log(1+x) to prevent log(0) errors

# Use IQR to filter out outliers
conditions = []
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR  # Use standard IQR multiplier (1.5)
    upper_bound = Q3 + 1.5 * IQR
    conditions.append((df[column] >= lower_bound) & (df[column] <= upper_bound))

# Apply all conditions at once to keep only non-outliers
df = df[np.logical_and.reduce(conditions)]

# Visualizing Distributions After Outlier Removal
plt.figure(figsize=(15, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.violinplot(x=df[column], color="blue")
    plt.title(f"Distribution of {column} After Outlier Removal")

plt.tight_layout()
plt.show()



````

**Insights from the Violin Plots (After Outlier Removal)**:

**Runtime**:

1.Most movies and shows have runtimes clustered around 60-120 minutes.
The distribution is fairly symmetrical after removing extreme durations.
IMDB Score:

2.Scores are concentrated between 5 and 7, suggesting most titles receive average ratings.
Few very low or very high-rated entries remain after outlier removal.
IMDB Votes (Log-Transformed):

3.The distribution is now smoother and closer to normal.
Most titles receive a modest number of votes, with a long tail of highly popular titles.
TMDB Popularity (Log-Transformed):

4.Similar to IMDB votes, popularity is right-skewed but better normalized after the log transformation.
Most titles have low-to-moderate popularity scores.

5.TMDB Score:
Scores are tightly clustered between 5 and 7, showing a bias toward mid-range ratings.
